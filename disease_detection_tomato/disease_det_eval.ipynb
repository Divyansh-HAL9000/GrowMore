```python
import tensorflow as tf
import tensorflow.keras as keras
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
models_path = "/gdrive/MyDrive/Colab Notebooks/models/"
test_imgs_path = "/gdrive/MyDrive/Colab Notebooks/leaf_images/"
```

```python
model97 = keras.models.load_model(models_path + "0.97test_accuracy") 
#model88 = keras.models.load_model(models_path + "0.880val_accuracy")
```

```python
images = ["1", "2", "3", "4"]
imgs = [f"{test_imgs_path}{image}.png" for image in images]
print(imgs)
```

```python
for img in imgs:
    image = cv2.imread(img)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image = cv2.resize(image, (224, 224))
    image2 = np.expand_dims(image, axis=0)
    plt.figure()
    plt.imshow(image)
#    plt.xlabel(image2.shape)
    classes = ["Bacterial Disease", "Fungal Disease", "Viral Disease", "Parasitic Disease (mites)", "Healthy"]
    plt.xlabel(f"""
        model prediction : {classes[np.argmax(model97.predict(image2))]}
    """)
```
